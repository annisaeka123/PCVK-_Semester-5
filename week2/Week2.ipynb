4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black
image dengan tipe data int16.

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img, pt1=(384,0), pt2=(510,128), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan
pt2 yang tertulis pada kode program.

```python
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius
yang tertulis pada kode program.

```python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai
berikut.

```python
cv.circle(black_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

Pembuatan black image kembali dilakukan dengan tipe data int32

```python
black_img2=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
vertices
```

Array tersebut kemudian di reshape sebagai berikut

```python
pts = vertices.reshape((-1,1,2))        #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, dan B
pts
```

Penambahan polyline pada black image kedua yang telah dibuat.

```python
cv.polylines(black_img2, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black_img2)
```

2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

```python
RB = img_rgb.copy()
RB[:,:,1] = 0     # channel hijau = 0 (sisa Red & Blue)

GB = img_rgb.copy()
GB[:,:,0] = 0     # channel merah = 0 (sisa Green & Blue)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(RB)
plt.title("Red + Blue")

plt.subplot(1,2,2)
plt.imshow(GB)
plt.title("Green + Blue")
plt.show()
```

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!

```python
img_boxes = img_rgb.copy()
h, w, _ = img_boxes.shape

for i in range(5):
    # koordinat random
    x1, y1 = np.random.randint(0, w-50), np.random.randint(0, h-50)
    x2, y2 = x1 + np.random.randint(20,100), y1 + np.random.randint(20,100)
    color = (np.random.randint(0,255), np.random.randint(0,255), np.random.randint(0,255))  # warna random
    cv.rectangle(img_boxes, (x1,y1), (x2,y2), color, -1)  # -1 untuk isi kotak penuh

plt.imshow(img_boxes)
plt.show()
```

6. Tampilkan image dengan posisi terbalik!

```python
flip_img = cv.flip(img_boxes, -1)   # -1 artinya flip horizontal & vertical
plt.imshow(flip_img)
plt.show()
```
