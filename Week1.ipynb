```python
#STEP 1
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #for image display
from skimage import io
from PIL import Image
import matplotlib.pylab as plt

```

```python
#STEP 2 (MEMBACA DAN MENAMPILKAN IMAGE)
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://i.pinimg.com/736x/26/84/03/268403e52e764afae9feaec8d4c1ba63.jpg"]
#baca dan tampilkan image
#loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                              #read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)     #resize image to half
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)      #convert color to RGB
    final_frame = cv.hconcat((image, image_2))          #concantenate image
    cv2_imshow(final_frame)                             #show image
    print('\n')
```

```python
#STEP 3 (MELIHAT UKURAN FILE IMAGE)
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

```python
#STEP 4 (MENGAKSES PIXEL DENGAN MEMBERIKAN GARIS HORIXONTAL BERWARNA PUTIH DI TENGAH IMAGE)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# TUGAS PRAKTIKUM D2
# 2. Garis Vertikal dan Garis menyilang diagonal
# a. Garis Vertikal
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertikal ditengah image
for x in range (tinggi):
  image_4[x,int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)
```

```python
#  b. garis silang diagonal
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi, lebar, _ = image.shape

# Membuat garis silang diagonal
for x in range(lebar):
    y1 = int((tinggi/lebar) * x)               # diagonal kiri atas -> kanan bawah
    y2 = tinggi - 1 - int((tinggi/lebar) * x)  # diagonal kiri bawah -> kanan atas

    image_5[y1, x] = [255, 255, 255]
    image_5[y2, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
```

```python
# 3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

tinggi, lebar, _ = image.shape

# Buat garis horizontal di tengah dengan panjang tertentu
panjang = 100
start_x = lebar//2 - panjang//2
end_x   = lebar//2 + panjang//2
y = tinggi//2

for x in range(start_x, end_x):
    image_6[y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame)
```

```python
# 4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

kotak_size = 100
start_x, start_y = 150, 100

for y in range(start_y, start_y+kotak_size):
    for x in range(start_x, start_x+kotak_size):
        if y < tinggi and x < lebar:                # agar tidak keluar batas
            image_7[y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_7))
cv2_imshow(final_frame)
```
